```python
#1 write a function is_prime(n) that returns true if a number is prime.otherwise false.
# def is_prime(n):
#     if (n // n & n ==1 ):
#         print(True)
#     elif(n // n & n==0):
#         print(False)
# print(is_prime(29))   # Ye saval maine cheat kiya hai iska Copilot approach ye hai niche

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):  # √n तक check
        if n % i == 0:   # remainder 0 मिला → divisible
            return False
    return True

print(is_prime(29))  # Output: True

```

```python
#2Create a function count_vovels[Test] that returns the total number of vovels in a string.\
def vovels(t):
    vov="aeiouAEIOU"
    vovels_list = []
    non_vovels_list = []
    for i in t:
        if i in vov:
            vovels_list.append(i)
        else:
            non_vovels_list.append(i)
    print("Vovels:" , vovels_list)
    print("Number of vovels:", len(vovels_list))
    print("Non-vovels:", non_vovels_list)
    print("Number of non-vovels:",len(non_vovels_list))

    return len(vovels_list),len(non_vovels_list)
     
print(vovels("HelloMyNameIs"))


```
