```python
s = [12,34,89,35,98,44,66,38,22,86,33,87,23]

odd_numbers = []
even_numbers = []

for num in s:
    if num % 2 == 0:   # remainder 0 → even
        even_numbers.append(num)
    else:              # remainder 1 → odd
        odd_numbers.append(num)

print("Odd Numbers:", odd_numbers)
print("Even Numbers:", even_numbers)

print("\nSquares of Odd Numbers:")
for num in odd_numbers:
    print(f"{num}^2 = {num**2}")

print("\nCubes of Odd Numbers:")
for num in odd_numbers:
    print(f"{num}^3 = {num**3}")

print("\nSquares of Even Numbers:")
for num in even_numbers:
    print(f"{num}^2 = {num**2}")

print("\nCubes of Even Numbers:")
for num in even_numbers:
    print(f"{num}^3 = {num**3}")
```
