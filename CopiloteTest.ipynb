```python
n = int(input("Enter your Number To check that it is Odd/ or EVEN:/")) 
if  n % 2  == 0 :
    print("Even Number: " )
else:
    print("Odd number")
print()
```

```python
#- User से age input लीजिए। अगर age 18 या उससे ज़्यादा है तो print करे "Eligible for Voting", वरना "Not Eligible"
age = int(input("Enter you age: To check That you are eligibal to vote or not: /  ")) 

if age >=18:
    print("Your are Eligible To Vote")
else:
    due= 18 -age
    print(f"Your are not eligible to vote waite for the {due} year after you are eligible")
```

```python
#- User से कोई number input लीजिए। अगर number positive है तो "Positive", negative है तो "Negative", और अगर 0 है तो "Zero" print करे
numb= int(input("Enter Number to check is it positive or negetive "))
if numb > 0:
    print("Positive Number ")
elif numb < 0 :
    print("Negative Number")
else:
    print("Zero")
```

मैं आपको Python में **if-else** के 3 beginner level और 3 intermediate level practice questions दे रहा हूँ।  

---

## 🟢 Beginner Level (आसान)
1. **Odd-Even Check**  
   एक प्रोग्राम लिखिए जो user से कोई number ले और बताए कि वह **even** है या **odd**।  

2. **Age Eligibility**  
   User से age input लीजिए। अगर age 18 या उससे ज़्यादा है तो print करे `"Eligible for Voting"`, वरना `"Not Eligible"`.  

3. **Positive-Negative Check**  
   User से कोई number input लीजिए। अगर number positive है तो `"Positive"`, negative है तो `"Negative"`, और अगर 0 है तो `"Zero"` print करे।  

---

## 🔵 Intermediate Level (थोड़ा challenging)
1. **Grading System**  
   User से marks input लीजिए (0–100)।  
   - अगर marks ≥ 90 → `"Grade A"`  
   - अगर marks ≥ 75 → `"Grade B"`  
   - अगर marks ≥ 50 → `"Grade C"`  
   - वरना `"Fail"`  

2. **Leap Year Check**  
   User से कोई year input लीजिए। अगर year 400 से divisible है या (4 से divisible है और 100 से divisible नहीं है) तो `"Leap Year"` print करे, वरना `"Not Leap Year"`।  

3. **Simple Calculator**  
   User से दो numbers और एक operator (`+`, `-`, `*`, `/`) input लीजिए।  
   - अगर operator `+` है तो addition करे  
   - अगर `-` है तो subtraction करे  
   - अगर `*` है तो multiplication करे  
   - अगर `/` है तो division करे  
   - वरना `"Invalid Operator"` print करे।
