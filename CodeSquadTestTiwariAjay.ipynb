```python
#1)  Write a program to count number of consonant in your name.
 

name = "ajay tiwari"    
name = name.lower()                
count = 0
vowels = "aeiou"
for ch in name:
    if ch.isalpha() and ch not in vowels:
        count += 1
print("Number of consonants in your name is :", count)
```

```python
#4 .write a program to print prime number b/w 200 to 500
for num in range(200, 501):    
    if num > 1:                
        for i in range(2, int(num**0.5) + 1):    
            if num % i == 0:  
                break
        else:
            print(num, end=" ")
```
