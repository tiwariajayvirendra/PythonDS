```python
#write a function is_prime(n) that returns true if a number is prime.otherwise false.
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
print(is_prime(3))
    
```

```python
#2Create a function count_vovels[Test] that returns the total number of vovels in a string.
def count_vowels(s):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in s if ch in vowels)

print(count_vowels("Test"))   

```

```python
#3Write a function factorial(n) using recursion.
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n-1)

print(factorial(5))   
```

```python
#4Create a function max_of_three(a,b,c)that returns tht largest number.
def max_of_three(a, b, c):
    return max(a, b, c)

print(max_of_three(3, 7, 5)) 
```

```python
#5Wap reverse_string(s)without using slicing([::-1]).
def reverse_string(s):
    result = ""
    for ch in s:
        result = ch + result
    return result

print(reverse_string("hello"))   
```

```python
#6Use lambda with map[]to squaes all numbers in List.
nums = [1, 2, 3, 4]
squares = list(map(lambda x: x**2, nums))
print(squares)   
```

```python
#7Use lambda with filter[]to extract even numbers from a list.
nums = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, nums))
print(evens)   
```

```python
#8use lambda with sorted []to sort a list of tuples based on the second element.
pairs = [(1, 3), (2, 1), (4, 2)]
sorted_pairs = sorted(pairs, key=lambda x: x[1])
print(sorted_pairs)   
```

```python
#9Write  alambda function to check whether a number is divisible by 3 and 5.
is_divisible = lambda x: x % 3 == 0 and x % 5 == 0
print(is_divisible(15))  
print(is_divisible(10))   
```

```python
#10use lambda wit reduce()to find the produ t of all elemeents in a list 
from functools import reduce
nums = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, nums)
print(product)  
```
