```python
#Create a class Google Pay In which information of diffrent account holders are given also information such as
#there UPI Bin , There bank Account Name, Total balance , are given , I need to access balance,and bank name of any indivisual account holder.
class GooglePay:
    def __init__(self, name, upi_bin, bank_name, balance):
        # Private attributes (encapsulation)
        self.name = name #public
        self.__upi_pin = upi_pin # Private
        self._bank_name = bank_name  # protected 
        self.__balance = balance   #Private

    # Getter methods to access private data
    def get_bank_name(self):
        return self._bank_name

    def get_balance(self):
        return self.__balance

    # Optional: method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"{amount} deposited successfully.")
        else:
            print("Invalid deposit amount.")

    # Optional: method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"{amount} withdrawn successfully.")
        else:
            print("Insufficient balance or invalid amount.")

# -------------------------------
# Example usage
account1 = GooglePay("Rahul", "UPI1234", "HDFC Bank", 5000)
account2 = GooglePay("Priya", "UPI5678", "ICICI Bank", 8000)

print("Rahul's Bank:", account1.get_bank_name())
print("Rahul's Balance:", account1.get_balance())

print("Priya's Bank:", account2.get_bank_name())
print("Priya's Balance:", account2.get_balance())

# Deposit and Withdraw
account1.deposit(2000)
print("Rahul's Updated Balance:", account1.get_balance())

account2.withdraw(3000)
print("Priya's Updated Balance:", account2.get_balance())
    
```

```python
#Create A class employee and Manger  wher in class Employee attribute like name and salaries given, where as in manager name ,salary, and department_name is given you need to access managename, salary and department ,
# Base Class
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

# Derived Class (Inheritance)
class Manager(Employee):
    def __init__(self, name, salary, department_name):
        # Employee class ke constructor ko call karna
        super().__init__(name, salary)
        self.department_name = department_name

# Object Creation
manager1 = Manager("Tiwari Ji", 50000, "Backend Development")

# Accessing Manager details
print("Manager Name:", manager1.name)
print("Manager Salary:", manager1.salary)
print("Manager Department:", manager1.department_name)
```
