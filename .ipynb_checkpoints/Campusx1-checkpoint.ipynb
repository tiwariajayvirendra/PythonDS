```python
#Gussing Game
#generate a random interger between 1 and 100
import random
jackpot = random.randint(1,100)
guss = int(input("guss karo"))
counter = 1

while guss != jackpot:
    if guss < jackpot:
        print('galat!guess higher')
    else:
        print("galat ! guess lower")
    guss = int(input("guss karo"))
    counter += 1
else:
    print("correct guss")
    print("attempts",counter)
```

```python
#Program - The current population of a town is 10000. the Population of the town is increasing at the rate of 10% per year. You have to write a program to find out the popultion at the end of each of the year
curr_pop = 10000
for i in range(5,0,-1):
    print(i,curr_pop)
    curr_pop = curr_pop -0.1*curr_pop
```
