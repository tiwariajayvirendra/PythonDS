```python
#
# If-else in Python 
email = input("enter email")
password = input('enter password')
if email == "ajaytripathi@gmail.com" and password == "1234":
    
    print("Welcome")
elif email == "ajaytripathi@gmail.com" and password != "1234":
    
    #Tell the user
    print("Incorrect password")
    password = input('enter Password')
    if password == "1234":
        print("Welcome, finally!")
    else:
        print("beta tumse no ho payega!")                     
else:
    print("Not Correct")
```
