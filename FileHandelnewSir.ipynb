```python
with open("schooldata.txt","a+") as file:
    name=input("enter name")
    age=int(input("age"))
    course=input("Your course")
    li=[name,age,course]
    file.write(str(li)+"\n")
```

```python
import ast
with open ("schooldata.txt")as file:
    data=file.readlines()
    for row in data:
        
        li=ast.literal_eval(row)
        print("Name:",li[0])
        print("Age:",li[1])
        print("course",li[2])
        print("------------------------------")
    
```
