```python
import json

# User se input lena
name = input("Enter Your Name: ")
age = input("Enter Your Age: ")
city = input("Enter Your City: ")

# Dictionary banani
data = {
    "name": name,
    "age": age,
    "city": city
}

# JSON file me save karna
with open("data.json", "w") as file:
    json.dump(data, file, indent=4)

print("Data saved successfully in data.json!")
```
