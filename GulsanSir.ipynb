```python
def permutations (nums, index=0):
    if index == len(nums): # Base case
        print(nums)
    else:
        for i in range (index,len(nums)):
            nums[index],nums[i]=nums[i], nums[index] #Swap
            permutations(nums,index +1)
            nums[index],nums[i] = nums[i],nums[index] #Backtrack
permutations([1,2,3])
```

```python
def power (a,b):
    if b == 0:
        return 1
    else:
        return a* power(a,b-1)
print(power(25,3))
```

```python
def reverse_string(s):
    if len(s) == 0: #Base case
        return " "
    else:
        return s[-1] +reverse_string(s[:-1])
print(reverse_string("Tiwari"))
```

```python
def is_palindrome(s):
    if len(s)<=1:
        return True
    elif s[0] != s[-1]:
        return False
    else:
        return is_palindrome(s[1:-1])
print(is_palindrome("radar"))
print(is_palindrome("radar"))
```

```python
def  binary_search(arr, target, left, right):
    if left > right: 
        return -1
    mid = (left + right) //2
    if arr [mid] == target:
        return mid
    elif target < arr [mid]:
        return binary_search(arr,target, left, mid-1)
    else:
        return binary_search(arr, target, mid+1, right)
arr = [1,3,5,7,9]
print(binary_search(arr , 9,0,len(arr)-1))
```

```python
def tower_of_hanoi(n,source,target,auxiliary):
    if n == 1:
        print(f"move disk 1 from {source} to {target}") # move n-1 disks to auxilaiary
    else:
        tower_of_hanoi(n -1,source, auxiliary, target)
        print(f"Move disc {n} from {source} to {target}")
        tower_of_hanoi(n-1,auxiliary,target,source)
tower_of_hanoi(3,'A','C','B')
```
