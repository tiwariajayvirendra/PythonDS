```python
#Write a function that takes in a list of integers and returns True if it Contains 007 in order 

# def spy_game(s):
#     if n in s.len:
#         prinet 
# def spy_game(t):
#     s = [0,0,7]
#     st =0

#     for i in t:
#         if i == s[st]:
#             st += 1
#             if st == len(s):
#                 print(True)
#             else:
#                 print(False)

# print(spy_game([1,0,2,3,4,0,7,5]))
def spy_game(nums):
    code = [0, 0, 7]   # target sequence
    code_index = 0     # track current position in code
    
    for num in nums:
        if num == code[code_index]:
            code_index += 1
            if code_index == len(code):  # agar pura sequence match ho gaya
                return True
    return False

# Test cases
print(spy_game([1,2,4,0,0,7,5]))  # True
print(spy_game([1,0,2,4,0,5,7]))  # True
print(spy_game([1,7,2,0,4,5,0]))  # False
```
