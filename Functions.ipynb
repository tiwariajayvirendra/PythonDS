```python
#Wap to print counting of each n every characater from a strin using dictionary n fucntion 
# Function to count characters
def char_count(s):
    d = {}  # empty dictionary
    for ch in s:
        if ch in d:
            d[ch] += 1   # increment count if already present
        else:
            d[ch] = 1    # add new character with count 1
    return d

# Example string
text = "hello world"

# Call the function
result = char_count(text)

# Print result
print("Character counts:")
for k, v in result.items():
    print(f"{k} : {v}")
```
