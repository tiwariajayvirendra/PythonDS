```python
#Lambda Function with MAP = will work on lIst tuple and string where it is iterable but not with Dict.
el= [1,2,3,4,5]
el1=list(map(lambda x: x*2,el))
print(el1)

 
```

```python
# lambda Function using Filter : Filter is  also itrable , works same as map but it works on the conditions ,
#if element is satisfied with the condion

el2=list(filter(lambda x : x%2==0,el))
print(el2)
```

```python
#Lambda Fucntion with Reduce : Reduce() means to return a single 
from functools import reduce
sum_all=reduce(lambda x,y:x+y,[10,20,30,40])
print(sum_all)
```

```python
# Question of Recursions:
#Sum of n natural numbers
def sumN(n):
    if(n==0): #base case 
        return 0
    sumNm = sumN(n-1)
    sum = sumNm+n
    return sum
print(sumN(5))
        
```
